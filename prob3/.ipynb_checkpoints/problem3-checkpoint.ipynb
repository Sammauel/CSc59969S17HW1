```python
# 77  
apples = np.array([df.iloc[78][7], df1.iloc[78][23], df2.iloc[78][39], df3.iloc[78][55]])
# 79
bananas = np.array([df.iloc[80][7], df1.iloc[80][23], df2.iloc[80][39], df3.iloc[80][55]])
# 80
berries = np.array([df.iloc[81][7], df1.iloc[81][23], df2.iloc[81][39], df3.iloc[81][55]])
# 81
grapes = np.array([df.iloc[82][7], df1.iloc[82][23], df2.iloc[82][39], df3.iloc[82][55]])
# 82
melons = np.array([df.iloc[83][7], df1.iloc[83][23], df2.iloc[83][39], df3.iloc[83][55]])
# 83 - skip the other oranges
oranges = np.array([df.iloc[84][7], df1.iloc[84][23], df2.iloc[84][39], df3.iloc[84][55]])
# 
citrus = np.array([df.iloc[87][7], df1.iloc[87][23], df2.iloc[87][39], df3.iloc[87][55]])

stoneFruit = np.array([df.iloc[88][7], df1.iloc[88][23], df2.iloc[88][39], df3.iloc[88][55]])

tropicalFruit = np.array([df.iloc[89][7], df1.iloc[89][23], df2.iloc[89][39], df3.iloc[89][55]])
```

```python
fruitsMen = np.array([0,1,2,3])
yearRange = ["94-98", "03-04", "05-06", "07-08"]
fig, ax = plt.subplots()
plt.xticks(fruitsMen,yearRange)
ax.plot(fruitsMen, apples, 'red', label = 'Apples as fruit', linestyle = '-')
ax.plot(fruitsMen, bananas, 'yellow', label = 'Bananas', linestyle = '--')
ax.plot(fruitsMen, berries, 'blue', label = 'Berries', linestyle = '-.')
ax.plot(fruitsMen, grapes, 'purple', label = 'Grapes', linestyle = ':')
ax.plot(fruitsMen, melons, 'tan', label = 'Melons', linestyle = '-')
ax.plot(fruitsMen, oranges, 'orange', label = 'Oranges, Total', linestyle = '--')
ax.plot(fruitsMen, citrus, 'black', label = 'Other citrus fruit', linestyle = '-.')
ax.plot(fruitsMen, stoneFruit, 'lime', label = 'Stone fruit', linestyle = ':')
ax.plot(fruitsMen, tropicalFruit, 'green', label = 'Tropical Fruit', linestyle = '-')
ax.set_xlabel('Year : 1994 - 2008')
ax.set_ylabel('Consumptions : Mean %')
ax.set_title('Fruit types over time (Men)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Judging from the results of the excel file it looks like the consumption rate of all the fruits has been slowly increasing over the decade with some fruits slighlty going down during the years 2004 to 2005.  The only significant change would be the sharp decrease in "Other citrus fruit" consumpution.  Men love eating every fruit but citrus.
